## 課題1

以下の図を再現してください

- 年齢の分布を、異なる階級幅を使って、4つのサブプロットをわけて描画する
   - 階級幅を適当で決めてよい
- 各サブプロットには、
    - x軸ラベル、y軸ラベル、サブプロット番号(a-d)を指定する
    - 水平方向のグリッド線を追加する
    - ヒストグラムの色を`lightblue`に変更する

```python
from matplotlib import pyplot as plt

# Figureオブジェクトを作成
fig = plt.figure(figsize=(6, 6))

# FigureにAxesを追加
ax_a = fig.add_subplot(2, 2, 1)  # 1行2列の最初のサブプロット
ax_b= fig.add_subplot(2, 2, 2)  # 1行2列の2番目のサブプロット
ax_c = fig.add_subplot(2, 2, 3)  # 1行2列の最初のサブプロット
ax_d = fig.add_subplot(2, 2, 4)  # 1行2列の2番目のサブプロット

# 年齢の分布を、異なる階級幅を使って、4つのサブプロットをわけて描画する
ax_a.hist(df["age"], bins=10, color = 'lightblue', label = 'Age')
ax_a.set_xlabel('Age')
ax_a.set_ylabel('People')
ax_a.grid(axis="y")
ax_b.hist(df["age"], bins=25, color = 'lightblue', label = 'Age')
ax_b.set_xlabel('Age')
ax_b.set_ylabel('People')
ax_b.grid(axis="y")
ax_c.hist(df["age"], bins=50, color = 'lightblue', label = 'Age')
ax_c.set_xlabel('Age')
ax_c.set_ylabel('People')
ax_c.grid(axis="y")
ax_d.hist(df["age"], bins=100, color = 'lightblue', label = 'Age')
ax_d.set_xlabel('Age')
ax_d.set_ylabel('People')
ax_d.grid(axis="y")

# レイアウトを調整
plt.tight_layout()

# グラフを表示
plt.show()
```

## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

![](./Figure2.png)

```python
# Figureオブジェクトを作成
fig = plt.figure(figsize=(6, 4))
fig.suptitle("Distribution of survivors by age and sex") 
# FigureにAxesを追加
ax_1 = fig.add_subplot(1, 2, 1)  # 1行2列の最初のサブプロット
ax_2= fig.add_subplot(1, 2, 2)  # 1行2列の2番目のサブプロット
# 年齢の分布を、諸条件に基づいて、2つのサブプロットをわけて描画する
width = 0.4 #ヒストグラムの幅
ax_1.hist(df.loc[df["sex"]=="male", ["age"]], alpha=0.5, label = 'male')
ax_1.hist(df.loc[df["sex"]=="female", ["age"]], alpha=0.5, label = 'female')
ax_1.legend()
ax_1.set_xlabel('Age')
ax_1.set_ylabel('People')

ax_2.hist(df["age"], color="gray", label = 'all')
ax_2.legend()
ax_2.set_xlabel('Age')
ax_2.set_ylabel('People')

ax_1.grid(axis="y")
ax_2.grid(axis="y")

# レイアウトを調整
plt.tight_layout()

# グラフを表示
plt.show()
```
