## 課題1

以下の図を再現してください

- 年齢の分布を、異なる階級幅を使って、4つのサブプロットをわけて描画する
   - 階級幅を適当で決めてよい
- 各サブプロットには、
    - x軸ラベル、y軸ラベル、サブプロット番号(a-d)を指定する
    - 水平方向のグリッド線を追加する
    - ヒストグラムの色を`lightblue`に変更する

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("/Users/weiyueshu/Desktop/titanic.csv")
df = df.dropna(subset=['Age'])
df['Age'] = pd.to_numeric(df['Age'])

male_ages = df[df['Sex'] == 'male']['Age'].values
female_ages = df[df['Sex'] == 'female']['Age'].values

fig, axs = plt.subplots(2, 2, figsize=(15, 10))  

axs[0, 0].hist(male_ages, bins=range(0, 81, 10), color='lightblue', alpha=0.7)
axs[0, 0].set_title('Age Distribution of Male Passengers (a)')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Count')
axs[0, 0].grid(axis='x', linestyle='--', alpha=0.7)

axs[0, 1].hist(female_ages, bins=range(0, 81, 10), color='lightblue', alpha=0.7)
axs[0, 1].set_title('Age Distribution of Female Passengers (b)')
axs[0, 1].set_xlabel('Age')
axs[0, 1].set_ylabel('Count')
axs[0, 1].grid(axis='x', linestyle='--', alpha=0.7)

axs[1, 0].hist(df['Age'], bins=range(0, 81, 10), color='lightblue', alpha=0.7)
axs[1, 0].set_title('Age Distribution of All Passengers (c)')
axs[1, 0].set_xlabel('Age')
axs[1, 0].set_ylabel('Count')
axs[1, 0].grid(axis='x', linestyle='--', alpha=0.7)

third_class_ages = df[(df['Pclass'] == 3) & (df['Age'].notnull())]['Age'].values
axs[1, 1].hist(third_class_ages, bins=range(0, 81, 10), color='lightblue', alpha=0.7)
axs[1, 1].set_title('Age Distribution of 3rd Class Passengers (d)')
axs[1, 1].set_xlabel('Age')
axs[1, 1].set_ylabel('Count')
axs[1, 1].grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

![](./Figure2.png)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("/Users/weiyueshu/Desktop/titanic.csv")
df = df.dropna(subset=['Age'])
df['Age'] = pd.to_numeric(df['Age'])

male_ages = df[df['Sex'] == 'male']['Age'].values
female_ages = df[df['Sex'] == 'female']['Age'].values
all_ages = df['Age'].values  

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))  

ax1.hist(male_ages, bins=np.arange(0, 81, 1), color='blue', alpha=0.7, label='Male')
ax1.hist(all_ages, bins=np.arange(0, 81, 1), color='grey', alpha=0.5, label='All Passengers', rwidth=0.5)
ax1.set_title('Age Distribution of Male Passengers')
ax1.set_xlabel('Age')
ax1.set_ylabel('Count')
ax1.legend()

ax2.hist(female_ages, bins=np.arange(0, 81, 1), color='red', alpha=0.7, label='Female')
ax2.hist(all_ages, bins=np.arange(0, 81, 1), color='grey', alpha=0.5, label='All Passengers', rwidth=0.5)
ax2.set_title('Age Distribution of Female Passengers')
ax2.set_xlabel('Age')
ax2.set_ylabel('Count')
ax2.legend()

plt.tight_layout()
plt.show()
```
