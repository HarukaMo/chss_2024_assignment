## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

```python
df2 = pd.DataFrame() 
df2 = df.loc[(df["sex"]=="male")]
df2

df3 = pd.DataFrame() 
df3 = df.loc[(df["sex"]=="female")]
df3
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,6))


age2 = df2["age"]
axs[0].hist(age, bins=50, color ="gray",  label="すべての乗客")
axs[0].hist(age2, bins=50, color ="blue",  label="男性乗客")
axs[0].set_title('男性乗客')
axs[0].set_xlabel('乗客の年齢(歳)', fontsize=14,labelpad=10)
axs[0].set_ylabel('人数', fontsize=14,labelpad=10)
axs[0].tick_params(labelsize=17)
axs[0].grid(axis="y")
axs[0].legend(fontsize=15)


age3 = df3["age"]
axs[1].hist(age, bins=50, color ="gray",  label="すべての乗客")
axs[1].hist(age3, bins=50, color ="red",  label="女性乗客")
axs[1].set_title('女性乗客')
axs[1].set_xlabel('乗客の年齢(歳)', fontsize=14,labelpad=10)
axs[1].set_ylabel('人数', fontsize=14, labelpad=10)
axs[1].tick_params(labelsize=17)
axs[1].grid(axis="y")
axs[1].legend(fontsize=15)

plt.tight_layout()
plt.show()
```
