## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

![](./Figure2.png)

```python
fig,axs=plt.subplots(1,2,figsize=(22,7))
age=df["age"]
age_male=df[df["sex"]=="male"]["age"]
age_female=df[df["sex"]=="female"]["age"]

axs[0].hist(age,bins=60,color="gray",label="すべての乗客")
axs[0].hist(age_male,bins=60,color="blue",label="男性乗客")
axs[0].set_xlabel('乗客の年齢(歳)',fontsize=20)
axs[0].set_ylabel('人数',fontsize=20)
axs[0].set_title('男性乗客',fontsize=25,loc="center")
axs[0].grid(axis="y")
axs[0].legend(["すべての乗客","男性乗客"],loc="upper right",fontsize=20)

axs[1].hist(age,bins=60,color="gray",label="すべての乗客")
axs[1].hist(age_female,bins=60,color="red",label="女性乗客")
axs[1].set_xlabel('乗客の年齢(歳)',fontsize=20)
axs[1].set_ylabel('人数',fontsize=20)
axs[1].set_title('女性乗客',fontsize=25,loc="center")
axs[1].grid(axis="y")
axs[1].legend(["すべての乗客","女性乗客"],loc="upper right",fontsize=20)

plt.show()
```
