```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Hiragino Maru Gothic Pro', 'Yu Gothic', 'Meirio', 'Takao', 'IPAexGothic', 'IPAPGothic', 'Noto Sans CJK JP']

fig, axs = plt.subplots(2,2, figsize=(12,8))

axs[0,0].hist(df['age'], bins=50, rwidth=0.8, color='skyblue', alpha=0.7)
axs[0,0].set_xlabel('年齢（歳）', fontsize=14)
axs[0,0].set_ylabel('人数', fontsize=14)
axs[0,0].set_title('a', fontsize=16, loc='left')
axs[0,0].xaxis.set_ticks_position('both')
axs[0,0].yaxis.set_ticks_position('both')
axs[0,0].grid(axis='y')
axs[0,0].set_yticks(range(0, 81, 20))
axs[0,0].minorticks_on()
axs[0,0].tick_params(direction='in', which='both')


axs[0,1].hist(df['age'], bins=30, rwidth=0.8, color='skyblue', alpha=0.7)
axs[0,1].set_xlabel('年齢（歳）', fontsize=14)
axs[0,1].set_ylabel('人数', fontsize=14)
axs[0,1].set_title('b', fontsize=16, loc='left')
axs[0,1].xaxis.set_ticks_position('both')
axs[0,1].yaxis.set_ticks_position('both')
axs[0,1].grid(axis='y')
axs[0,1].set_yticks(range(0, 121, 20))
axs[0,1].minorticks_on()
axs[0,1].tick_params(direction='in', which='both')

axs[1,0].hist(df['age'], bins=15, rwidth=0.8, color='skyblue', alpha=0.7)
axs[1,0].set_xlabel('年齢（歳）', fontsize=14)
axs[1,0].set_ylabel('人数', fontsize=14)
axs[1,0].set_title('c', fontsize=16, loc='left')
axs[1,0].xaxis.set_ticks_position('both')
axs[1,0].yaxis.set_ticks_position('both')
axs[1,0].grid(axis='y')
axs[1,0].set_yticks(range(0, 151, 50))
axs[1,0].minorticks_on()
axs[1,0].tick_params(direction='in', which='both')

axs[1,1].hist(df['age'], bins=10, rwidth=0.8, color='skyblue', alpha=0.7)
axs[1,1].set_xlabel('年齢（歳）', fontsize=14)
axs[1,1].set_ylabel('人数', fontsize=14)
axs[1,1].set_title('d', fontsize=16, loc='left')
axs[1,1].xaxis.set_ticks_position('both')
axs[1,1].yaxis.set_ticks_position('both')
axs[1,1].grid(axis='y')
axs[1,1].set_yticks(range(0, 251, 50))
axs[1,1].minorticks_on()
axs[1,1].tick_params(direction='in', which='both')

fig.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```

## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

![](./Figure2.png)

```python
df = df.dropna(subset=['age'])

male_ages = df[df['sex'] == 'male']['age']
female_ages = df[df['sex'] == 'female']['age']
all_ages = df['age']

fig, axs = plt.subplots(1, 2, figsize=(14, 5))
axs[0].hist(all_ages, bins=50, color='gray', label='すべての乗客')
axs[0].hist(male_ages, bins=50, color='blue', alpha=0.7, label='男性乗客')
axs[0].set_title('男性乗客', fontsize=14)
axs[0].set_xlabel('乗客の年齢(歳)',fontsize=14)
axs[0].set_ylabel('人数', fontsize=14)
axs[0].xaxis.set_ticks_position('both')
axs[0].yaxis.set_ticks_position('both')
axs[0].grid(axis='y')
axs[0].minorticks_on()
axs[0].tick_params(direction='in', which='both')
axs[0].legend(fontsize=14, frameon=False)

axs[1].hist(all_ages, bins=50, color='gray', label='すべての乗客')
axs[1].hist(female_ages, bins=50, color='red', alpha=0.7, label='女性乗客')
axs[1].set_title('女性乗客', fontsize=14)
axs[1].set_xlabel('乗客の年齢(歳)',fontsize=14)
axs[1].set_ylabel('人数', fontsize=14)
axs[1].xaxis.set_ticks_position('both')
axs[1].yaxis.set_ticks_position('both')
axs[1].grid(axis='y')
axs[1].minorticks_on()
axs[1].tick_params(direction='in', which='both')
axs[1].legend(fontsize=14, frameon=False)

plt.tight_layout()
plt.show()
```
